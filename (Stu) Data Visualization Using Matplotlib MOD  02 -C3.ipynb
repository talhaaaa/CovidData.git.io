## COVID-19 Dataset Description

The dataset spans from January 4 to July 31 and contains daily COVID-19 reports for various countries. Initially, only data from China is present due to the virus being limited to China during that period.

Each row in the dataset provides the following information:

- **Date_reported**: The date when the data was reported in year-month-day format.
- **Country**: The country to which the report corresponds.
- **New_cases**: The number of new cases reported for that date and country.
- **Cumulative_cases**: The total number of cases up to that date in that country.
- **New_deaths**: The number of new deaths reported for that date and country.
- **Cumulative_deaths**: The total number of deaths up to that date in that country.

The dataset is organized in chronological order, beginning from January 4 and extending to July 31. This sequence of data points, ordered by time, is referred to as a time series.


```python
df["Date_reported"]=pd.to_datetime(df['Date_reported'],format = "%Y-%m-%d")
```

```python
df.info()
```

# Italy

```python
import matplotlib.pyplot as plt
Italy=df["Country"] == "Italy"
Italy_df=df[Italy]
Italy_df
```

## Graph

**Note:** When dealing with time series data, line graphs are the most commonly used visualization.

In a time series line graph:
- The x-axis represents the independent time values.
- The y-axis represents the dependent variable being observed.


```python
plt.plot (Italy_df["Date_reported"],Italy_df['Cumulative_cases']) #x-axis , y-axis
# heading
plt.title("Italy :Cumulative Cases ")

# x label

plt.xlabel("Date")

# y label

plt.ylabel("Number of cases")

# to hide hex codes

plt.show()

# interference

print("Generaally a quantity which increase very quickly in\ the begining and then show down is called logarithmic graph")
```

# India 

```python
def plot_reported_cases(country_name):
    
    country_df = df[df['Country'] == country_name]


    plt.plot (country_df["Date_reported"],country_df['Cumulative_cases']) #x-axis , y-axis
    # heading
    plt.title(country_name+ "Cumulative Cases ")

    # x label

    plt.xlabel("Date")

    # y label

    plt.ylabel("Number of cases")

    # to hide hex codes

    plt.show()

plot_reported_cases("India")
# interference

print("Generaally a quantity that has slow start but \ go hight afterwards is called exponential graph")
```

# Poland

```python
plot_reported_cases("Poland")
```

# Plotting Multiple Countries

```python
#plotting 1st 10 unique countries
len(df['Country'].unique())
Countries = df["Country"].unique()
Countries[:10]
```

```python
for i in Countries[:10]:
    #print(i)
    plot_reported_cases(i)
```
